```python
import networkx as nx
import numpy as np
import pprint

xs = [0, 5, 16, 20, 33, 23, 35, 25, 10]
ys = [15, 20, 24, 20, 25, 11, 7, 0, 3]
mgraph = [[np.abs(xs[i]-xs[j])+np.abs(ys[i]-ys[j]) for j in range(len(ys))] for i in range(len(xs))]
# pprint.pprint(mgraph)
elist = []
for i in range(len(xs)):
    for j in range(i):
        elist.append((str(i), str(j), mgraph[i][j]))

#pprint.pprint(elist)
G = nx.Graph()
G.add_weighted_edges_from(elist)
min_spantree = nx.algorithms.tree.mst.minimum_spanning_tree(G, weight='weight', algorithm='kruskal')
#nx.draw(G, pos=nx.spring_layout(G))  # use spring layout
nx.draw(min_spantree, pos=nx.spring_layout(min_spantree))  # use spring layout
#print(min_spantree.edges())
edge_weight = []
for edge in min_spantree.edges():
    edge_weight.append(mgraph[int(edge[0])][int(edge[1])])
#print(edge_weight)
#print(np.sum(edge_weight))
```
