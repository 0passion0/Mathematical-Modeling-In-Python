```python
from scipy.optimize import fsolve
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

x = np.linspace(-10, 10, 40)
y = np.sin(x) + np.random.normal(loc=0, scale=0.05, size=x.shape)
xx = np.linspace(-10, 10, 1000)

fig = plt.figure(figsize=(15, 5))
plt.scatter(x, y, c='r', label='Raw Data')
# kind是表示几次样条好像.
f = interp1d(x, y, kind=3)
plt.plot(xx, f(xx), c='b', label='Interpolated Curve')
plt.plot(xx, np.sin(xx), c='g', label='Standard Curve')
plt.ylabel('x')
plt.legend()
plt.tight_layout()
plt.show()
```
